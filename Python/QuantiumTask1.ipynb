# Splitting every words of products.

```python
unique_products = transaction_data["PROD_NAME"].unique();

allwords = [];

for product in unique_products:
    allwords.extend(product.split());

words_df = pd.DataFrame(allwords , columns=["Words"]);
print(words_df.tail(6));
```

# Remove Numeric Characters


```python
words_df = words_df[~words_df["Words"].str.contains(r"\d")];
print(words_df.tail(5));
```

# Remove Special Characters like &.


```python
words_df = words_df[words_df["Words"].str.isalpha()];
print(words_df.tail(6));
```

```python
#lowercasing every words
words_df  = words_df["Words"].str.lower();
print(words_df.tail());
```

```python
count_salsa = (words_df.values == "salsa").sum();
count_dip = (words_df.values == "dip").sum();
count_dips = (words_df.values == "dips").sum();
count_sauce = (words_df.values == "sauce").sum();
print(count_salsa);
print(count_dip);
print(count_dips);
print(count_sauce);
```

# Removing Salsa

```python

transaction_data['SALSA'] = transaction_data['PROD_NAME'].str.lower().str.contains(r'\bsalsa\b');

before_count = transaction_data['PROD_NAME'].str.lower().str.contains(r'\bsalsa\b').sum();
print("Rows containing salsa before removal:", before_count);

transaction_data = transaction_data[transaction_data['SALSA'] == False];

#Removing unnecessary column named salsa after operation.
transaction_data = transaction_data.drop(columns=['SALSA']);

after_count = transaction_data['PROD_NAME'].str.lower().str.contains(r'\bsalsa\b').sum();
print("Rows containing salsa after removal:", after_count);
```

# T-Testing.


```python

data["UNIT_PRICE"] = data["TOT_SALES"] / data["PROD_QTY"]

unit_price_per_customer = data.groupby(
    ["LYLTY_CARD_NBR", "LIFESTAGE", "PREMIUM_CUSTOMER"]
)["UNIT_PRICE"].mean().reset_index()

mainstream = unit_price_per_customer[
    (unit_price_per_customer["PREMIUM_CUSTOMER"] == "Mainstream") &
    (unit_price_per_customer["LIFESTAGE"].isin([
        "YOUNG SINGLES/COUPLES",
        "MIDAGE SINGLES/COUPLES"
    ]))
]["UNIT_PRICE"]

budget_premium = unit_price_per_customer[
    (unit_price_per_customer["PREMIUM_CUSTOMER"].isin(["Budget", "Premium"])) &
    (unit_price_per_customer["LIFESTAGE"].isin([
        "YOUNG SINGLES/COUPLES",
        "MIDAGE SINGLES/COUPLES"
    ]))
]["UNIT_PRICE"]

from scipy.stats import ttest_ind

t_stat, p_value = ttest_ind(mainstream, budget_premium, equal_var=False)

print("T-statistic:", round(t_stat, 4))
print("P-value:", round(p_value, 6))
```

# BRAND prefrence comparison.


```python

target = data[
    (data["PREMIUM_CUSTOMER"] == "Mainstream") &
    (data["LIFESTAGE"] == "YOUNG SINGLES/COUPLES")
]

others = data[
    ~(
        (data["PREMIUM_CUSTOMER"] == "Mainstream") &
        (data["LIFESTAGE"] == "YOUNG SINGLES/COUPLES")
     )
]

target_brand = target["BRANDS"].value_counts(normalize=True) #Normalize is used to convert into percentage.

others_brand = others["BRANDS"].value_counts(normalize=True)

brand_comparison = pd.DataFrame({
    "Target_Segment": target_brand,
    "Others": others_brand
}).fillna(0)

brand_comparison["Difference in %"] = (
    brand_comparison["Target_Segment"] -
    brand_comparison["Others"]
)

brand_comparison.sort_values("Difference in %", ascending=False)
```
